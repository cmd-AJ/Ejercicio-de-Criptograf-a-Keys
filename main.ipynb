## Inciso 1

Realizar la creación de un script que permita generar llaves dinámicas (utilizar ASCII)

```python
import random

# Importo las letras del abecedario y los digitos osea del 0 al 9 y tambien las puntuaciones o caracteres especiales
import string


# length de valores llaves osea la cantidad de caracteres en una llave
length = 10



def generar_llave(longitud=16):
    # Conjunto de caracteres ASCII
    caracteres = (
        string.ascii_letters + 
        string.digits +        
        string.punctuation    
    )

    
    # Generar llave aleatoria con el random entonces elige aleatorio el caracter que la cadena caracteres
    llave = ''
    for i in range (longitud):
        llave = llave +  random.choice(caracteres)
    return llave

# imprimo mi respuesta de la llave
print("La llave generada es:", generar_llave(length))
```

```python
length = 10
llave = generar_llave(length)

llave = list(llave)

palabra = "mazariegos"

palabra = list(palabra)


# se saco del ejercicio anterior el de convertir ascii a binario
def decimal_a_binario(n):
    # si es cero es cero 
    if n == 0:
        return "0"
    binario = ""
    # https://www.geeksforgeeks.org/python/python-program-to-covert-decimal-to-binary-number/
    # Se saco de geek for greeks pero el tema es que se divide entre 2 hasta conseguir el residuo  
    while n > 0:
        binario = str(n % 2) + binario
        n //= 2
    return binario


for e in range(0,length):
    llave[e] = ord(llave[e])
    llave[e] = decimal_a_binario(llave[e])

# tomar nota hacer luego en xor para cifrar el texto

for e in range(0,len(palabra)):
    palabra[e] = ord(palabra[e])
    palabra[e] = decimal_a_binario(palabra[e])



llave = ", ".join(llave)
palabra = ", ".join(palabra)


def xor_binario(a, b):

    # EN caso de aplicar xor dinamico se utiliza se encuentra primero el valor de digitos más alto 
    max_len = max(len(a), len(b))
    a = a.zfill(max_len)
    b = b.zfill(max_len)    

    # se convierte en array para clasificar cada digito para luego utilizarlo en un for e implementar a cada uno el xor por cada digito
    bin1 = list(a)
    bin2 = list(b)

    result = ""


    for i in range(0,max_len):
        if bin1[i] != bin2[i]:
            result = result + "1"  # cuando 0 xor 1 = 1
        else:
            result = result + "0" # cuando 0 xor 0 ó 1 xor 1 = 0 
    
    return result


result = xor_binario(llave,palabra )

print(result)


```
